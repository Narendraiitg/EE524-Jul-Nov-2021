```python
import numpy as np ## importing the numpy library
import matplotlib.pyplot as plt
fig,ax=plt.subplots()
a1=-2
b1=2
a2=3
b2=5
a3=2
b3=5
x=np.linspace(a1-1,b1+1,1000);
y=np.linspace(a2-1,b2+1,1000);
z=np.linspace(a3-1,b3+1,1000);
def uniform_cdf(x,a1,b1): ## defining the function
    if x < a1:
        return 0;
    elif a1<=x<=b1:
        return (x-a1)/(b1-a1);
    elif x>=b1:
        return 1;
def uniform_cdf(y,a2,b2): ## defining the function
    if y < a2:
        return 0;
    elif a2<=y<=b2:
        return (y-a2)/(b2-a2);
    elif y>=b2:
        return 1;
def uniform_cdf(z,a3,b3): ## defining the function
    if z < a3:
        return 0;
    elif a3<=z<=b3:
        return (z-a3)/(b3-a3);
    elif z>=b3:
        return 1;    
ax.plot(x,[uniform_cdf(k,a1,b1)for k in x]);
ax.plot(y,[uniform_cdf(l,a2,b2)for l in y]);
ax.plot(z,[uniform_cdf(m,a3,b3)for m in z]);
ax.set_title("CDF Of Uniform Distribution with parameters a & b")
ax.set_xlabel("x")
ax.set_ylabel("value of f(x)")
ax.axhline(color = "Black");
ax.axvline(color = "Black");
```

# Discrete distributions 

(a) Consider a transmitter that can transmit messages only in the form
of 2 bits 0 and 1. The receiver receives a single message sent by the
transmitter say X. Since, we do not know before hand which bit
was sent X follows a certain distribution. Which distribution is this?
Find the PMF and plot it. Also calculate the mean and variance of
X. Assume probability for receiving 0(failure) is p = 0.1, 0.5, 0.8.

Ans: The distribution in above question is bernouli trial since there is only one trial as receiver receives only a single message by the transmitter & it have only two outcomes 0 & 1
P(success) = P = 0.1, 0.5, 0.8
P(Faluire) = 1-P = 0.9, 0.5, 0.2
Mean of X  = P = 0.1, 0.5, 0.8 
Variance of X  = P(1-P) = 0.09, 0.25, 0.16

```python
import numpy as numpy
import matplotlib.pyplot as plt
x=[0,1] # an array
P=0.1
def pmf(x,P): # defining the function
    if x==0:  # condition for return
        return P;
    elif x==1:
        return 1-P;
    else:
        return 0;

for i in range(0,2): # providing the range to for loop
    plt.plot(x[i],pmf(x[i],P),'bs') # plotting the pmf by calling the function
plt.xlim(-1,5) # limit of x axis
plt.title("PMF for P=0.1") # providing the title to the plot
plt.ylabel("Bernouli trial") 
plt.xlabel("x axis")
```
